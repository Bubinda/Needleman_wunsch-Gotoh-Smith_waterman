```python
def print_matrix(mat):
    for i in range(0, len(mat)):
        print("[", end = "")
        for j in range(0, len(mat[i])):
            print(mat[i][j], end = "")
            if j != len(mat[i]) - 1:
                print("\t", end = "")
        print("]\n")

```

```python
#NW
gap_penalty = -2
match_award = 6
mismatch_penalty = -1

seq1 = "LNSGYHFCGGSLI"
seq2 = "LSMGCGGALY"


def match_score(alpha, beta):
    if alpha == beta:
        return match_award
    elif alpha == '-' or beta == '-':
        return gap_penalty
    else:
        return mismatch_penalty

def needleman_wunsch(seq1, seq2):
    
    n = len(seq1)
    m = len(seq2)  
    
    score = np.zeros((m+1,n+1), dtype=np.int64)
    
    for i in range(0, m + 1):
        score[i][0] = gap_penalty * i
    
    for j in range(0, n + 1):
        score[0][j] = gap_penalty * j
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            match = score[i - 1][j - 1] + match_score(seq1[j-1], seq2[i-1])
            delete = score[i - 1][j] + gap_penalty
            insert = score[i][j - 1] + gap_penalty
            score[i][j] = max(match, delete, insert)

    return score

print_matrix(needleman_wunsch(seq1, seq2))
```

```python
#NW
def needleman_wunsch_path_finder(seq1, seq2):
    
    n = len(seq1)  
    m = len(seq2)
    
    score = np.zeros((m+1,n+1), dtype=np.int64)
   
    for i in range(0, m + 1):
        score[i][0] = gap_penalty * i
    
    for j in range(0, n + 1):
        score[0][j] = gap_penalty * j
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            match = score[i - 1][j - 1] + match_score(seq1[j-1], seq2[i-1])
            delete = score[i - 1][j] + gap_penalty
            insert = score[i][j - 1] + gap_penalty
            score[i][j] = max(match, delete, insert)

    align1 = ""
    align2 = ""
    path = []
    
    i = m
    j = n
    
    while i > 0 and j > 0: 
        score_current = score[i][j]
        score_diagonal = score[i-1][j-1]
        score_up = score[i][j-1]
        score_left = score[i-1][j]

        if score_current == score_diagonal + match_score(seq1[j-1], seq2[i-1]):
            align1 += seq1[j-1]
            align2 += seq2[i-1]
            i -= 1
            j -= 1
            
        elif score_current == score_up + gap_penalty:
            align1 += seq1[j-1]
            align2 += '-'
            j -= 1
        elif score_current == score_left + gap_penalty:
            align1 += '-'
            align2 += seq2[i-1]
            i -= 1
        path.append((i,j))
    while j > 0:
        align1 += seq1[j-1]
        align2 += '-'
        j -= 1
    while i > 0:
        align1 += '-'
        align2 += seq2[i-1]
        i -= 1
    
    align1 = align1[::-1]
    align2 = align2[::-1]
    
    return(align1, align2, path)

output1, output2, pathway = needleman_wunsch_path_finder(seq1, seq2)

print(output1 + "\n" + output2)
print('optimal path:' , pathway)
    
   
```

```python
#Gotoh
open_gap_penalty = -6
extend_gap_penalty = -1
match_award = 6
mismatch_penalty = -1

seq1 = "LNSGYHFCGGSLI"
seq2 = "LSMGCGGALY"


def match_score(alpha, beta):
   return match_award if alpha == beta else mismatch_penalty

def g(number):
  return open_gap_penalty + extend_gap_penalty * number


def Gotoh(seq1, seq2):
    
    n = len(seq1)
    m = len(seq2)  
    
    matrix_D = np.zeros((n+1,m+1)) #, dtype=np.int64
    matrix_P = np.zeros((n+1,m+1))
    matrix_Q = np.zeros((n+1,m+1))

    for i in range(0, n):
        matrix_D[i+1][0] = g(i+1)
        matrix_P[i][0] =  np.nan
        matrix_Q[i+1][0] = -np.inf
    
    for j in range(0, m):
        matrix_D[0][j+1] = g(j+1)
        matrix_P[0][j+1] = -np.inf
        matrix_Q[0][j] = np.nan
 

    #compute matrix P (vertical)
    for i in range(1, n + 1):
      for j in range(1, m + 1):
        mat_d = matrix_D[i - 1][j] + g(1)
        mat_p = matrix_P[i - 1][j] + extend_gap_penalty
        matrix_P[i][j] = max(mat_d,mat_p)


    #compute matrix Q (horizontal)
    for i in range(1, n + 1):
      for j in range(1, m + 1):
        mat_d = matrix_D[i][j-1] + g(1)
        mat_q = matrix_Q[i][j-1] + extend_gap_penalty
        matrix_Q[i][j] = max(mat_d,mat_q)

   #compute matrix D (diagonal)
    for i in range(1, n + 1):
        for j in range(1, m + 1):
          mat_d = matrix_D[i - 1][j - 1] + match_score(seq1[i-1], seq2[j-1])
          mat_p = matrix_P[i][j]
          mat_q = matrix_Q[i][j]
          matrix_D[i][j] = max(mat_d, mat_p,mat_q)

    return matrix_D, matrix_P, matrix_Q
```

```python
d,p,q = Gotoh(seq1, seq2)

print("this is the D matrix")
print_matrix(d)
print("this is the P matrix")
print_matrix(p)
print("this is the q matrix")
print_matrix(q)
```

```python
#SMW
gap_penalty = -2
match_award = 6
mismatch_penalty = -1

seq1 = "LNSGYHFCGGSLI"
seq2 = "LSMGCGGALY"


def match_score(alpha, beta):
    if alpha == beta:
        return match_award
    elif alpha == '-' or beta == '-':
        return gap_penalty
    else:
        return mismatch_penalty

def smith_waterman(seq1, seq2):
    
    n = len(seq1)
    m = len(seq2)  
    
    score = np.zeros((m+1,n+1), dtype=np.int64)
    
    for i in range(0, m + 1):
        score[i][0] = 0
    
    for j in range(0, n + 1):
        score[0][j] = 0
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            match = score[i - 1][j - 1] + match_score(seq1[j-1], seq2[i-1])
            delete = score[i - 1][j] + gap_penalty
            insert = score[i][j - 1] + gap_penalty
            score[i][j] = max(match, delete, insert,0)

    return score

print_matrix(smith_waterman(seq1, seq2))
```

```python
#SMW
def smith_waterman_path_finder(seq1, seq2):
    
    n = len(seq1)  
    m = len(seq2)
    align1 = ""
    align2 = ""
    path = []

    score = np.zeros((m+1,n+1), dtype=np.int64)
   
    for i in range(0, m + 1):
        score[i][0] = 0
    
    for j in range(0, n + 1):
        score[0][j] = 0
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            match = score[i - 1][j - 1] + match_score(seq1[j-1], seq2[i-1])
            delete = score[i - 1][j] + gap_penalty
            insert = score[i][j - 1] + gap_penalty
            score[i][j] = max(match, delete, insert,0)

 

    maximum = np.unravel_index(np.argmax(score,axis=None), score.shape)
    i = maximum[0]
    j = maximum[1]
    path.append(maximum)

    while i > 0 and j > 0:
        if score[i][j] == 0:
          break;
        else:
          score_current = score[i][j]
          score_diagonal = score[i-1][j-1]
          score_up = score[i][j-1]
          score_left = score[i-1][j]

          if score_current == score_diagonal + match_score(seq1[j-1], seq2[i-1]):
              align1 += seq1[j-1]
              align2 += seq2[i-1]
              i -= 1
              j -= 1
              
          elif score_current == score_up + gap_penalty:
              align1 += seq1[j-1]
              align2 += '-'
              j -= 1
          elif score_current == score_left + gap_penalty:
              align1 += '-'
              align2 += seq2[i-1]
              i -= 1
          path.append((i,j))
    while j > 0:
        align1 += seq1[j-1]
        align2 += '-'
        j -= 1
    while i > 0:
        align1 += '-'
        align2 += seq2[i-1]
        i -= 1
    
    align1 = align1[::-1]
    align2 = align2[::-1]
    
    return(align1, align2, path)

output1, output2, pathway = smith_waterman_path_finder(seq1, seq2)

print(output1 + "\n" + output2)
print('optimal path:' , pathway)
```
